Today's challenge Python Numpy

 NumPy Arrays & Operations
 Task:
Create 1D, 2D, 3D arrays.Perform addition, subtraction, multiplication, division.
Use functions: mean(), sum(), std(), reshape(), arange(), linspace().
Find indices of max/min values.

```python
import numpy as np

# -------------------------------
# 1. Create 1D, 2D, 3D Arrays
# -------------------------------

# 1D array
arr1 = np.array([10, 20, 30, 40, 50])

# 2D array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

# 3D array
arr3 = np.array([ [[1, 2], [3, 4]],
                  [[5, 6], [7, 8]] ])

print("1D Array:\n", arr1)
print("\n2D Array:\n", arr2)
print("\n3D Array:\n", arr3)

# -------------------------------
# 2. Perform Arithmetic Operations
# -------------------------------

print("\nAddition (arr1 + 5):", arr1 + 5)
print("Subtraction (arr1 - 5):", arr1 - 5)
print("Multiplication (arr1 * 2):", arr1 * 2)
print("Division (arr1 / 5):", arr1 / 5)

# -------------------------------
# 3. Use Functions: mean(), sum(), std()
# -------------------------------

print("\nMean of arr1:", np.mean(arr1))
print("Sum of arr1:", np.sum(arr1))
print("Standard Deviation of arr1:", np.std(arr1))

# -------------------------------
# 4. reshape(), arange(), linspace()
# -------------------------------

# arange(): numbers with step
a = np.arange(1, 11)      # 1 to 10
print("\nArange Array:\n", a)

# reshape: convert 1D → 2D
reshaped = a.reshape(2, 5)
print("\nReshaped Array (2x5):\n", reshaped)

# linspace(): equal spaced values
b = np.linspace(0, 1, 5)   # 5 values between 0 and 1
print("\nLinspace Array:\n", b)

# -------------------------------
# 5. Find indices of Min & Max
# -------------------------------

print("\nIndex of Max Value (arr1):", np.argmax(arr1))
print("Index of Min Value (arr1):", np.argmin(arr1))
```

 Task:

```python
import pandas as pd

# ----------------------------------------
# 1. Import Dataset into Pandas DataFrame
# ----------------------------------------

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
df = pd.read_csv(url)

print("Initial DataFrame:")
print(df.head())

# ----------------------------------------
# 2. Explore Series & Columns
# ----------------------------------------

print("\nColumn Names:")
print(df.columns)

print("\nAccessing a Single Column (Country):")
print(df['Country'].head())

print("\nAccessing Multiple Columns:")
print(df[['Country', 'Region']].head())

# ----------------------------------------
# 3. Add, Update, Delete Columns / Rows
# ----------------------------------------

# Add new column
df['Country_Length'] = df['Country'].apply(len)

# Update a column value
df.loc[0, 'Country'] = "USA (Edited)"

# Delete a column
df = df.drop(columns=['Country_Length'])

# Delete a row
df = df.drop(index=1)

print("\nAfter Add/Update/Delete:")
print(df.head())

# ----------------------------------------
# 4. Set Custom Index and Sort
# ----------------------------------------

# Set custom index
df = df.set_index('Country')

# Sort by Region
df_sorted = df.sort_values(by='Region')

print("\nSorted DataFrame by Region:")
print(df_sorted.head())
```

Today's challenge:- Data Exploration & Wrangling
 Task:
Explore dataset: head(), tail(), info(), describe(), value_counts().


Detect & handle missing values: dropna(), fillna().


Merge, concat, join datasets.


Group by and aggregate data.
 Project Connection: Perform data cleaning, initial wrangling, and summary statistics.




```python
import pandas as pd

# Load dataset from online source
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
df = pd.read_csv(url)

# -------------------------------------------------------
# 1. Explore Dataset (head, tail, info, describe, counts)
# -------------------------------------------------------

print("HEAD:\n", df.head(), "\n")
print("TAIL:\n", df.tail(), "\n")
print("INFO:")
print(df.info(), "\n")
print("DESCRIBE:\n", df.describe(include='all'), "\n")

print("VALUE COUNTS (Region):")
print(df["Region"].value_counts(), "\n")

# -------------------------------------------------------
# 2. Detect & Handle Missing Values (dropna, fillna)
# -------------------------------------------------------

print("Missing Values Count:\n", df.isna().sum(), "\n")

# Fill missing values
df_filled = df.fillna("Unknown")

# Drop rows with missing values
df_dropped = df.dropna()

print("After fillna():\n", df_filled.head(), "\n")
print("After dropna():\n", df_dropped.head(), "\n")

# -------------------------------------------------------
# 3. Merge, Concat, Join
# -------------------------------------------------------

extra = pd.DataFrame({
    "Region": ["Asia", "Europe"],
    "Code": [101, 102]
})

# Merge
merged = pd.merge(df, extra, on="Region", how="left")
print("MERGED:\n", merged.head(), "\n")

# Concat
concat_df = pd.concat([df.head(3), df.tail(3)], ignore_index=True)
print("CONCAT:\n", concat_df, "\n")

# Join (requires index)
df2 = df.set_index("Country")
df3 = pd.DataFrame({
    "GDP": [50000, 60000, 30000],
}, index=["USA", "Germany", "France"])

joined = df2.join(df3, how="left")
print("JOINED:\n", joined.head(), "\n")

# -------------------------------------------------------
# 4. Group By + Aggregate
# -------------------------------------------------------

grouped = df.groupby("Region")["Country"].count()
print("GROUPED BY REGION (count):\n", grouped, "\n")
```

Day 5 – Matplotlib Visualization
 Task:
Line chart, scatter plot, bar chart, pie chart.


Histogram & box plot to show distributions and outliers.
 Project Connection: Visualize trends, distributions, and outliers in your dataset.

```python
import matplotlib.pyplot as plt

# Simple numeric data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# ----------------------
# 1. Line chart
# ----------------------
plt.figure()
plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# ----------------------
# 2. Scatter Plot
# ----------------------
plt.figure()
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# ----------------------
# 3. Bar Chart
# ----------------------
plt.figure()
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("X")
plt.ylabel("Values")
plt.show()

# ----------------------
# 4. Pie Chart
# ----------------------
plt.figure()
plt.pie([10, 20, 30, 40], labels=["A", "B", "C", "D"], autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()

# ----------------------
# 5. Histogram
# ----------------------
plt.figure()
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data, bins=4)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# ----------------------
# 6. Box Plot
# ----------------------
plt.figure()
plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()
```

Today's challenge:-
 Seaborn & Plotly Advanced Visualization
 Task:
Heatmap, pair plots, violin, swarm, density plots.
Interactive Plotly charts (scatter, bar, line).
 Project Connection: Identify correlations and relationships; create interactive visuals.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load sample dataset
df = sns.load_dataset("tips")


plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

sns.pairplot(df, hue="sex")
plt.show()


plt.figure(figsize=(7,5))
sns.swarmplot(data=df, x="day", y="tip", hue="sex")
plt.title("Swarm Plot - Tips by Day")
plt.show()


plt.figure(figsize=(7,5))
sns.kdeplot(data=df, x="total_bill", hue="sex", fill=True)
plt.title("Density Plot - Total Bill by Gender")
plt.show()


fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="day",
    size="size",
    title="Interactive Scatter Plot - Bill vs Tip"
)
fig.show()


fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="sex",
    title="Average Total Bill per Day",
    barmode="group"
)
fig.show()


fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="sex",
    title="Average Total Bill per Day",
    barmode="group"
)
fig.show()


fig = px.line(
    df.sort_values("size"),
    x="size",
    y="total_bill",
    color="day",
    title="Interactive Line Chart - Total Bill vs Table Size"
)
fig.show()


```
