**Challenge 2.**
Tuple, Immutable vs Mutable, List Creation

 Project Title:
Retail Inventory Manager

 Scenario: You are managing an inventory system for a retail shop. Certain product categories are fixed, but the actual products can be added or removed by the staff.

Tasks:
•	Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
•	Create a list of products that can be updated dynamically.
•	Ask the user to enter a new product and add it to the list.
•	Ask the user to input the index of a product to remove it from the list.
•	Display the final list of products and the product categories.


```python
# Fixed product categories (Immutable → Tuple)
categories = ("Electronics", "Stationery", "Clothing")

# Dynamic product list (Mutable → List)
products = ["Laptop", "Pen", "Shirt"]

print("Initial Products:", products)
print("Product Categories:", categories)

# Add new product
new_product = input("Enter a new product to add: ")
products.append(new_product)
print("\nProduct added successfully!")

# Remove product by index
print("\nCurrent Products with Index:")
for i, item in enumerate(products):
    print(i, "→", item)

remove_index = int(input("Enter the index of the product to remove: "))

# Validate index
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)
    print(f"\n'{removed_item}' removed successfully!")
else:
    print("\nInvalid index! No product removed.")

# Final Output
print("\nFinal Product List:", products)
print("Product Categories (Tuple):", categories)
```

List Methods and Slicing

 Project Title: Contact Directory

 Scenario: You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.

 Tasks:
•	Create a 2D list where each row holds a contact’s name, phone number, and email address.
•	Allow the user to add a new contact.
•	Display all contacts sorted alphabetically by name.
•	Allow the user to search for contacts by name. The search should support partial matches.
•	After sorting, display only the first 3 contacts in the list.


```python
# Contact Management System

# Step 1: Create a 2D list for contacts [Name, Phone, Email]
contacts = [
    ["Arun", "9876543210", "arun@example.com"],
    ["Bala", "9123456780", "bala@example.com"],
    ["Chitra", "9988776655", "chitra@example.com"]
]

# Step 2: Allow user to add a new contact
name = input("Enter contact name: ")
phone = input("Enter phone number: ")
email = input("Enter email address: ")

contacts.append([name, phone, email])
print("\nContact added successfully!\n")

# Step 3: Sort contacts alphabetically by name
contacts.sort(key=lambda x: x[0].lower())

# Step 4: Display only the first 3 contacts after sorting
print("=== First 3 Contacts (Sorted by Name) ===")
for contact in contacts[:3]:
    print(f"Name: {contact[0]}, Phone: {contact[1]}, Email: {contact[2]}")

# Step 5: Allow user to search by (partial) name
search_name = input("\nEnter name to search: ").lower()

print("\n=== Search Results ===")
found = False
for contact in contacts:
    if search_name in contact[0].lower():
        print(f"Name: {contact[0]}, Phone: {contact[1]}, Email: {contact[2]}")
        found = True

if not found:
    print("No contact found matching your search.")
```

Dictionary, Set, Frozenset

 Project Title:
Student Grades & Hobbies Tracker

 Scenario:
A student activity coordinator is collecting information about student grades and hobbies.
They want to ensure grades are updated properly and hobby lists are stored efficiently.

Tasks:
 Create a dictionary with student names as keys and their grades as values.
Allow the user to add or update grades for at least 3 students.
Create a set of hobbies and allow the user to input at least 3 hobbies.
 Convert the set of hobbies into a frozenset and display it.


```python
# Step 1: Create a dictionary with student names and grades
students = {}

# Step 2: Allow the user to add/update grades for at least 3 students
for i in range(3):
    name = input("Enter Student Name: ")
    grade = input("Enter Grade: ")
    students[name] = grade

print("\nUpdated Student Grades Dictionary:")
print(students)

# Step 3: Create a set of hobbies and allow the user to input at least 3 hobbies
hobbies = set()
for i in range(3):
    hobby = input("Enter a hobby: ")
    hobbies.add(hobby)

print("\nHobby Set:")
print(hobbies)

# Step 4: Convert the set of hobbies into a frozenset and display
frozen_hobbies = frozenset(hobbies)

print("\nFrozen Hobby Set:")
print(frozen_hobbies)

```

Project Title: Shopping Cart System

Scenario: You are creating a simple billing system where customers can enter the prices of items they want to purchase. The program should keep accepting prices until the user decides to stop, then apply appropriate discounts before generating the final bill.

Tasks:
•	Continuously accept item prices using a while loop.

•	Stop taking input when the user enters 0 (sentinel value).

•	Keep track of the total number of items and the subtotal amount.

•	Ensure no negative prices are allowed (ask again without counting as an item).

•	Apply discounts based on the subtotal:

•	20% for totals ≥ 1000

•	10% for totals ≥ 500

•	5% for totals ≥ 200

•	0% for totals < 200

•	Display a clear bill summary including subtotal, discount amount, and final total.


```python
subtotal = 0
item_count = 0

print("Enter item prices (Enter 0 to finish):")

while True:
    price = float(input("Enter price: "))

    # Stop when user enters 0
    if price == 0:
        break

    # Check for negative price
    if price < 0:
        print("Negative price not allowed! Try again.")
        continue

    # Update counts
    subtotal += price
    item_count += 1

# Apply discount based on subtotal
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Display Summary
print("\n------ BILL SUMMARY ------")
print(f"Total Items Purchased: {item_count}")
print(f"Subtotal Amount: ₹{subtotal:.2f}")
print(f"Discount Applied: {discount_rate * 100:.0f}%")
print(f"Discount Amount: ₹{discount_amount:.2f}")
print(f"Final Total to Pay: ₹{final_total:.2f}")
print("-------------------------")
```

Topic: range, for loop, enumerate, list comprehension

Project Title: Employee Hours Report

 Scenario: You are analyzing student exam scores to generate performance reports.

 Tasks:

•	Create a list of 10 random exam scores ranging from 0 to 100 using list comprehension.

•	Display each score with its index using the enumerate function.

•	Find all scores greater than 75 and display them.

•	Calculate and display the average score.
import random


```python
import random
# 1️ Create a list of 10 random exam scores (0–100)
scores = [random.randint(0, 100) for _ in range(10)]

# Display the full score list
print("Exam Scores:", scores)

# 2️ Display each score with its index using enumerate
print("\nScore List with Index:")
for index, score in enumerate(scores):
    print(f"Student {index} → Score: {score}")

# 3️ Find all scores greater than 75
high_scores = [score for score in scores if score > 75]
print("\nScores > 75:", high_scores)

# 4️ Calculate and display the average score
average_score = sum(scores) / len(scores)
print("\nAverage Score:", round(average_score, 2))
```

Built-in Functions and User-Defined Functions

Project Title: Simple Calculator App Scenario: A company wants to provide users with a simple calculator to perform basic mathematical operations.

 Tasks:

•	Create functions for addition, subtraction, multiplication, and division.

•	Ask the user to input two numbers and the operation they want to perform.

•	Call the appropriate function and display the result.

•	Use built-in functions like abs(), round(), and pow() within your solution.

•	Handle division by zero gracefully and inform the user.


```python
# Function definitions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:  # handling division error
        return "Error: Division by zero is not allowed!"
    return a / b

# Asking user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Choose operation (+, -, *, /): ")

# Performing operation
if operation == '+':
    result = add(num1, num2)
elif operation == '-':
    result = subtract(num1, num2)
elif operation == '*':
    result = multiply(num1, num2)
elif operation == '/':
    result = divide(num1, num2)
else:
    result = "Invalid operation selected!"

# Displaying results with built-in functions usage
if isinstance(result, str):
    print(result)
else:
    print("Result:", result)
    print("Absolute value:", abs(result))
    print("Rounded value:", round(result))
    print("Power (result^2):", pow(result, 2))
```
